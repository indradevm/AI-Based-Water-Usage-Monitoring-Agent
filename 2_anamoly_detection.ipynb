```python
import pandas as pd
import numpy as np
```

```python
# Loading cleaned dataset from previous notebook
df = pd.read_csv(r"C:/Users/BIT/Desktop/AIML ASSIGNMENT/data/cleaned_daily_water_consumption.csv")

# Convert Date again (important after reload)
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

```python
# Extracting useful time features

df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Day_of_Week'] = df['Date'].dt.dayofweek  # 0 = Monday

df.head()
```

```python
# Rolling average (smooth trend)
df['Rolling_Mean_3'] = df['Water_Consumed_Liters'].rolling(window=3).mean()

# Rolling standard deviation (variation)
df['Rolling_Std_3'] = df['Water_Consumed_Liters'].rolling(window=3).std()
```

```python
# Removing initial rows with NaN due to rolling
df = df.dropna()

df.head()
```

```python
# Previous day's consumption
df['Lag_1'] = df['Water_Consumed_Liters'].shift(1)

# Removing NaN created due to lag
df = df.dropna()
```

```python
# Converting categorical variables to numeric

df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

df['Activity_Level'] = df['Activity_Level'].map({
    'Low': 0,
    'Medium': 1,
    'High': 2
})
```

```python
# One-hot encoding for City
# Helps model understand location differences
df = pd.get_dummies(df, columns=['City'], drop_first=True)
```

```python
# Displaying final columns
print("Final features:\n", df.columns)

# Shape of dataset
print("\nShape after feature engineering:", df.shape)
```

```python
# Saving dataset for next step (anomaly detection)

df.to_csv("C:/Users/BIT/Desktop/AIML ASSIGNMENT/data/feature_engineered_data.csv", index=False)

print("Feature engineered dataset saved successfully!")
```
